```python
from google.colab import files
import pandas as pd

# Upload the CSV file from your system
uploaded = files.upload()

# Read the CSV file (replace 'data.csv' with your filename)
df = pd.read_csv('data.csv')
```

```python
# Assuming you've already uploaded the CSV and created the dataframe 'df'
import pandas as pd
import matplotlib.pyplot as plt

# c. Show the basic statistical description about the data
basic_stats = df.describe()
print("Basic Statistical Description:\n", basic_stats)

# d. Check if the data has null values
null_values = df.isnull().sum()
print("\nNull Values in each column:\n", null_values)

# d(i). Replace the null values with the mean of their respective columns
df.fillna(df.mean(), inplace=True)
print("\nNull values replaced with mean.")

# e. Select at least two columns and aggregate data using: min, max, count, mean
aggregated_data = df[['Calories', 'Pulse']].agg(['min', 'max', 'count', 'mean'])
print("\nAggregated Data:\n", aggregated_data)

# f. Filter the dataframe to select the rows with calories values between 500 and 1000
filtered_df_500_1000 = df[(df['Calories'] >= 500) & (df['Calories'] <= 1000)]
print("\nRows with Calories between 500 and 1000:\n", filtered_df_500_1000)

# g. Filter the dataframe to select the rows with calories > 500 and pulse < 100
filtered_df_cal_gt_500_pulse_lt_100 = df[(df['Calories'] > 500) & (df['Pulse'] < 100)]
print("\nRows with Calories > 500 and Pulse < 100:\n", filtered_df_cal_gt_500_pulse_lt_100)

# h. Create a new dataframe 'df_modified' that excludes 'Maxpulse' column
df_modified = df.drop(columns=['Maxpulse'])
print("\nNew dataframe 'df_modified' without 'Maxpulse' column:\n", df_modified.head())

# i. Delete the 'Maxpulse' column from the main df dataframe
df.drop(columns=['Maxpulse'], inplace=True)
print("\n'Maxpulse' column deleted from original dataframe.")

# j. Convert the datatype of 'Calories' column to int datatype
df['Calories'] = df['Calories'].astype(int)
print("\nCalories column datatype converted to int.")

# k. Create a scatter plot for 'Duration' and 'Calories'
plt.scatter(df['Duration'], df['Calories'])
plt.title('Duration vs Calories Scatter Plot')
plt.xlabel('Duration')
plt.ylabel('Calories')
plt.grid(True)
plt.show()
```

```python
# Import Libraries
from google.colab import files
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Upload the file
uploaded = files.upload()

# Load Dataset into a DataFrame (you need to use the key of the uploaded file)
filename = list(uploaded.keys())[0]
data = pd.read_csv(filename)

# Split the data into training and testing sets (1/3 for testing)
X = data[['YearsExperience']]  # Features (independent variable)
y = data['Salary']  # Target (dependent variable)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=1/3, random_state=42)

# Train the model
model = LinearRegression()
model.fit(X_train, y_train)

# Predict on both train and test sets
y_train_pred = model.predict(X_train)
y_test_pred = model.predict(X_test)

# Calculate Mean Squared Error
mse = mean_squared_error(y_test, y_test_pred)
print(f"Mean Squared Error: {mse}")

# Visualize train and test data with scatter plots
plt.figure(figsize=(10, 5))

# Training data plot
plt.scatter(X_train, y_train, color='blue', label='Training data')
plt.plot(X_train, y_train_pred, color='black', label='Regression Line (Train)')

# Test data plot
plt.scatter(X_test, y_test, color='green', label='Test data')
plt.plot(X_test, y_test_pred, color='red', label='Regression Line (Test)')

plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.title('Linear Regression: Salary vs Experience')
plt.legend()
plt.show()

```
